### 5. Provide two examples of DataFrame filtering using logical conditions
- Example: Filtering by category, threshold, etc.

```python
# This filters songs numerically and by if they have a high popularity
popular = df_new.loc[df['Popularity'] > 80, 'Track'] 
#This filters songs categorically based on whether Madonna made them 
madonna_songs = df_new.loc[df['Artist'] == 'Madonna', 'Track']  
print("Songs with high popularity:\n", popular)
print("\nSongs made by Madonna:\n", madonna_songs)
```

Above the code provides a filtered list of song names included in the dataset that were either made by Maddonna or have a popularity of over 80.

### 8. Use .groupby() to compute summary statistics
