```python
#This code imports the Libraries needed to open the data frame and visualize it in graphs. 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#This will allow pandas to read and open the 1990s Hits Dataset
df= pd.read_csv('1990sClassicHits.csv')
df
```

```python
# Creates a copy of the original data frame to avoid modifying the original data set
df_new=df.copy()
# This creates a function to convert a duration from a string to an integer in seconds because I would like to analyze duration as an integer.  
def convert_dur(duration):
    minutes, seconds = map(int, duration.split(':')) # This separates the duration into minutes and seconds, converts each into integers separately. 
    return minutes * 60 + seconds # This returns the totals of minutes and seconds, and then converts the minutes to seconds to make the whole total in seconds.
# This will apply the function to the 'Duration' column in the new data frame.
df_new['Duration'] = df_new['Duration'].apply(convert_dur)
```

```python
df_new #This displays are new data frame
```

```python
# This is the number of songs per artist, get the top 10 artists with the most hits, then create a bar plot to visualize it.
df_new['Artist'].value_counts().head(10).plot(kind='bar', color='Red', edgecolor='Black', figsize=(6,4))
# This sets the title and axis labels of the chart.
plt.title("Top 10 Artist Hit Songs Count")
plt.xlabel("Artist")
plt.ylabel("Number of Hit Songs")
# This will make the limit of the y-axis go from 0-20 and rotate the names of the artist by 45 degrees. 
plt.ylim(0, 20)
plt.xticks(rotation=50)
plt.show() #This shows the bar chart. 
```

This bar chart shows the top 10 artists with the highest number of hit songs in the dataset. Each bar represents the number of hit songs associated with an artist. An observation I noticed is that there's a sharp drop in hit count from the top 3 artists to the others in the top 10. This suggests that there is a steep popularity curve in the artists, where only a few of the most popular dominated the charts in the 1990s and defined the decade. 

```python
# This sets the figure size for the plot.
plt.figure(figsize=(10, 5))
# Creates a histogram with a line curve to show the trend for the 'Danceability' column.
sns.histplot(df_new['Danceability'], bins=25, kde=True, color='blue', edgecolor='black')
# This sets the title and axis labels of the chart.
plt.title('Distribution of Danceability')
plt.xlabel('Danceability')
plt.ylabel('Frequency')
# This will limit the y-axis to 0-100 and limit the x-axis to 0-1.
plt.ylim(0,100)
plt.xlim(0,1)
# This adds a grid to the histogram and then displays the histogram. 
plt.grid(True)
plt.show()
```

- The highest bar and density peak happen between 0.6 and 0.8, showing that most songs in this dataset are very danceable and very popular during the 90s. 
- The histogram is skewed to the right, implying that most songs in the 1990s were designed to be upbeat and make you want to dance which makes sense because of the trend of pop and R&B music in the 1990s.

```python
# This will group the dataset by 'Track' and calculate the mean Danceability and Valence for each track.
dance_by_pop = df_new.groupby('Track')[['Danceability', 'Valence']].mean().sort_values(by='Danceability', ascending=False)
# This will take the top 10 tracks with the highest average Danceability.
top_dance= dance_by_pop.head(10)
# Plots a horizontal bar chart showing the Danceability and Valence for the top tracks.
top_dance.plot(kind='barh', figsize=(12, 6), color=['darkblue', 'maroon'], legend= True)
# This sets the title and axis labels of the chart.
plt.title('Top 10 Tracks by Average Danceability and Valence')
plt.xlabel('Average Score')
plt.ylabel('Tracks')
# This will display the legend in the best location, to make sure the chart is tightly shown for a neat presentation, then display the chart. 
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```


- Tracks with high Danceability often also show high Valence, implying that upbeat danceable songs also tend to sound happier.
- Some tracks  with slightly lower Valence in this top 10 may be highly danceable but carry a more neutral or even melancholic tone. 

```python
# This sets the figure size for the plot.
plt.figure(figsize=(15, 10))
# Creates a heat map for the correlation between all numerical columns in the DataFrame.
sns.heatmap(df_new.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt='.3f')
# This sets the title of the chart. 
plt.title('Correlation Matrix of Numerical Columns')
# Will rotate the columns on the x-axis by 45 degrees and then display the heatmap. 
plt.xticks(rotation=45)
plt.show()
```

- Valence has a high positive correlation with both Danceability and Energy, further showing that more energetic or danceable songs tend to feel happier or more positive.
- Loudness and Energy have a high positive correlation, showing that the louder a song is, the more it tends to have more energy.
- Acousticness has a high negative correlation with both Energy and Valence, showing that the more acoustic a song is, the more it sounds melancholic, and the energy of the song goes down.
